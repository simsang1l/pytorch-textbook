```python
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
from kmeans_pytorch import kmeans, kmeans_predict
```

```python
df = pd.read_csv("./data/iris.csv")
df.info()
print('-'*20)
print(df)
```

```python
data = pd.get_dummies(df, columns = ["Species"])
data
```

```python
from sklearn.model_selection import train_test_split

x, y = train_test_split(data, test_size = 0.2, random_state = 123)
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda')
else :
    device = torch.device('cpu')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit(data).transform(x)
y_scaled = scaler.fit(data).transform(y)
```

```python
x = torch.from_numpy(x_scaled)
y = torch.from_numpy(y_scaled)
```

```python
print(x.size())
print(y.size())
print(x)
```

```python
num_clusters = 3
cluster_ids_x, cluster_centers = kmeans(
    X = x, num_clusters = num_clusters, distance = 'euclidean', device = device
)
```

```python
print(cluster_ids_x)
print(cluster_centers)
```

```python
cluster_ids_y = kmeans_predict(
    y, cluster_centers, 'euclidean', device = device
)
```

```python
print(cluster_ids_y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (4, 3), dpi = 160)
plt.scatter(y[:, 0], y[:, 1], c= cluster_ids_y, cmap='viridis', marker='x')

plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1],
            c = 'white',
            alpha = 0.6,
            edgecolors='black',
            linewidths = 2
            )

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

x_train = np.load('./data/data.npy')

gmm = GaussianMixture(n_components = 2)
gmm.fit(x_train)

print(gmm.means_)
print('\n')
print(gmm.covariances_)

x, y = np.meshgrid(np.linspace(-1, 6), np.linspace(-1, 6))
xx = np.array([x.ravel(), y.ravel()]).T

z = gmm.score_samples(xx)
z = z.reshape((50, 50))

plt.contour(x, y, z)
plt.scatter(x_train[:, 0], x_train[:, 1])
plt.show()
```

```python
import numpy as np
from sklearn.datasets import load_digits
from minisom import MiniSom
from pylab import plot, axis, show, pcolor, colorbar, bone

digits = load_digits()
data = digits.data
labels = digits.target
```

```python
som = MiniSom(16, 16, 64, sigma = 1.0, learning_rate=0.5)
som.random_weights_init(data)
print("Som ì´ˆê¸°í™”")
som.train_random(data, 10000)
print("\n. SOM ì§„í–‰ ì¢…ë£Œ")

bone()
pcolor(som.distance_map().T)
colorbar()
```

```python
labels[labels=='0'] = 0
labels[labels=='1'] = 1
labels[labels=='2'] = 2
labels[labels=='3'] = 3
labels[labels=='4'] = 4
labels[labels=='5'] = 5
labels[labels=='6'] = 6
labels[labels=='7'] = 7
labels[labels=='8'] = 8
labels[labels=='9'] = 9

markers = ['o', 'v', '1', '3', '8', 's', 'p', 'x', 'D', '*']
colors = ['r', 'g', 'b', 'y', 'c', (0, 0.1, 0.8), (1, 0.5, 0), (1, 1, 0.3), 'm', (0.4, 0.6, 0)]
for cnt, xx in enumerate(data):
    w = som.winner(xx)
    plot(w[0]+.5, w[1]+.5, markers[labels[cnt]],
         markerfacecolor = "None", markeredgecolor=colors[labels[cnt]],
         markersize=12, markeredgewidth=2
         )
show()
```
